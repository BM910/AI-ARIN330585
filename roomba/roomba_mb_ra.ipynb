Máy hút bụi Model-based Reflex Agent với tập luật

Link Github: https://github.com/BM910/AI-ARIN330585

Ma trận 2 chiều "room" biểu diễn sàn nhà:
    0 - Không có bụi, có thể đi qua được.
    1 - Có bụi, có thể đi qua được.
    2 - Vật cản, không đi qua được.
    3 - Máy hút bụi.
    4 - Có bụi và có máy hút bụi tại vị trí đó.

Ma trận hai chiều "visited" ghi nhận những nơi máy hút bụi đã đi qua:
    0 - Chưa đi qua
    1 - Đã đi qua

1 - Máy hút bụi dùng get_moves() để tìm các hướng đi hợp lệ: UP, DOWN, LEFT, RIGHT
2 - Trong những hướng đi hợp lệ dùng prioritize_moves(moves) để tìm ra hướng đi ưu tiên:
    Ưu tiên thứ nhất: hướng đi vào ô có bụi (room[x][y] = 1)
    Ưu tiên thứ nhì: hướng đi vào ô chưa từng ghé thăm (visited[x][y] = 0)
3 - Nếu không chọn được hướng đi ưu tiên thì chọn một hướng ngẫu nhiên trong các hướng hợp lệ

```python
import numpy as np

room = [[3, 2, 1, 2, 1, 0],
        [0, 1, 0, 0, 0, 2],
        [2, 0, 1, 1, 0, 2],
        [2, 2, 2, 1, 0, 1]]

visited = []

for i in room:
    row = []
    for j in i:
        if j == 2:
            row.append(0)
        else:
            row.append(0)
    visited.append(row)

def find_roomba():
    global room
    for i in range(len(room)):
        for j in range(len(room[i])):
            if room[i][j] == 3 or room[i][j] == 4:
                return (i, j)

def get_moves():
    global room
    moves = []
    x, y = find_roomba()
    directions = [(-1, 0, "UP"), (1, 0, "DOWN"), (0, -1, "LEFT"), (0, 1, "RIGHT")]

    for dx, dy, name in directions:
        nx, ny = x + dx, y + dy
        if 0 <= nx < len(room) and 0 <= ny < len(room[0]) and room[nx][ny] != 2:
            moves.append((dx, dy, name))

    return moves

def execute_move(direction):
    x, y = find_roomba()
    dx, dy, name = direction
    room[x][y] = 0
    visited[x][y] = 1
    room[x + dx][y + dy] += 3
    # print_room(room)
    # print(f"Roomba moved {name}")

def prioritize_moves(moves):
    for move in moves:
        dx, dy, name = move
        x, y = find_roomba()
        nx = x + dx
        ny = y + dy
        if room[nx][ny] == 1:
            return move
        
    for move in moves:
        dx, dy, name = move
        x, y = find_roomba()
        nx = x + dx
        ny = y + dy
        if visited[nx][ny] == 0:
            return move

    n = np.random.randint(0, len(moves))
    return moves[n]

def print_full():
    global room, visited
    x, y = find_roomba()

    print("ROOM\t\tVISITED")
    for i in range(len(room)):
        for j in range(len(room[i])):
            print(room[i][j], end=" ")
        print(end="\t")
        for j in range(len(visited[i])):
            print(visited[i][j], end=" ")
        print()

def roomba(turns):
    global room

    print_full()

    for i in range(turns):
        x, y = find_roomba()
        if room[x][y] == 4:
            print(f"Roomba cleaned [{x}, {y}]")

        moves = get_moves()
        move = prioritize_moves(moves)
        execute_move(move)
        
        print("-" * 20)
        print_full()
        # print(f"Avaiable moves: {moves}")
        print(f"Roomba moved {move[2]}")

        if not any (1 in  row for row in room):
            print("Roomba has cleaned room")
            return

roomba(20)
```
