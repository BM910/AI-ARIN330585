Percept: Ma trận 2 chiều biểu diễn sàn nhà, trong đó:
    0 - Không có bụi, có thể đi qua được.
    1 - Có bụi, có thể đi qua được.
    2 - Vật cản, không đi qua được.
    3 - Máy hút bụi.
    4 - Có bụi và có máy hút bụi tại vị trí đó.

State (x, y) biểu diễn vị trí của máy hút bụi trên ma trận percept.

```python
import numpy as np

def move(percept, state):
    x, y = state
    x_dir = [-1, 0, 1, 0]
    y_dir = [0, 1, 0, -1]
    valid = False

    while not valid:
        n = np.random.randint(4)
        x_move = x_dir[n] + x
        y_move = y_dir[n] + y
        
        if x_move < 0 or x_move > 3 or y_move < 0 or y_move > 3:
            continue
        elif percept[x_move][y_move] == 2:
            continue
            
        valid = True
    
    percept[x][y] = 0
    percept[x_move][y_move] += 3 
    
    print(f"Máy hút bụi di chuyển tới ({x_move}, {y_move})")
    return (x_move, y_move)

def rule(percept, state):
    x, y = state
    if percept[x][y] == 3:
        state = move(percept, state)
    elif percept[x][y] == 4:
        print(f"Máy hút bụi dọn ({x}, {y})")
        percept[x][y] = 3
        state = move(percept, state)
    return state

def print_percept(percept):
    for row in percept:
        print(row)
    print("-" * 15)

percept = [
    [0, 0, 0, 0],
    [0, 2, 2, 0],
    [0, 1, 1, 1],
    [0, 0, 1, 0]
]

state = (0, 0)
percept[state[0]][state[1]] = 3

print("Sàn nhà:")
print_percept(percept)

for i in range(50):
    state = rule(percept, state)
    print_percept(percept)
```
