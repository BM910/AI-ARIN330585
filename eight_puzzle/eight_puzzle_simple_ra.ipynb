Giải câu đố eight_puzzle bằng simple reflex agent

```python
GOAL = [[1, 2, 3],
        [4, 5, 6],
        [7, 8, 0]]

state = [[0, 2, 3],
         [1, 7, 5],
         [4, 8, 6]]

def interpret_input(percept):
        for i in range(3):
                for j in range(3):
                        if percept[i][j] == 0:
                                return (i, j)

def print_state(state):
        print(state[0])
        print(state[1])
        print(state[2])

def simple_reflex_agent(step):
        global state
        done = 0
        while state != GOAL:
                zero_pos = interpret_input(state)
                i = zero_pos[0]
                j = zero_pos[1]
                goal_at_zero = GOAL[i][j]
                print("Current state:")
                if i - 1 >= 0 and state[i - 1][j] == goal_at_zero:
                        print_state(state)
                        state[i][j] = state[i - 1][j]
                        state[i - 1][j] = 0
                        print("NUMBER", state[i][j], "GO DOWN")
                        print_state(state)
                elif j + 1 < 3 and state[i][j + 1] == goal_at_zero:
                        print_state(state)
                        state[i][j] = state[i][j + 1]
                        state[i][j + 1] = 0
                        print("NUMBER", state[i][j], "GO LEFT")
                        print_state(state)
                elif i + 1 < 3 and state[i + 1][j] == goal_at_zero:
                        print_state(state)
                        state[i][j] = state[i + 1][j]
                        state[i + 1][j] = 0
                        print("NUMBER", state[i][j], "GO UP")
                        print_state(state)
                elif j - 1 >= 0 and state[i][j - 1] == goal_at_zero:
                        print_state(state)
                        state[i][j] = state[i][j - 1]
                        state[i][j - 1] = 0
                        print("NUMBER", state[i][j], "GO RIGHT")
                        print_state(state)
                done += 1
                if done >= step:
                        return

simple_reflex_agent(50)
```
