Giải câu đố Eight-puzzle Model-based Reflex Agent với thuật toán tìm kiếm mù BFS

Link Github: https://github.com/BM910/AI-ARIN330585

```python
from collections import deque

class Node:
    def __init__(self, state, parent=None, action=None):
        self.state = state
        self.parent = parent
        self.action = action
    
    def __repr__(self):
        return f"Node({self.state})"

class EightPuzzleAgent:
    def __init__(self, initial_state, goal_state):
        self.initial_state = tuple(initial_state)
        self.goal_state = tuple(goal_state)
        self.actions = ['UP', 'DOWN', 'LEFT', 'RIGHT']

    def get_blank_pos(self, state):
        return state.index(0)

    def result(self, state, action):
        new_state = list(state)
        idx = self.get_blank_pos(state)
        row, col = divmod(idx, 3)

        if action == 'UP' and row > 0:
            new_state[idx], new_state[idx-3] = new_state[idx-3], new_state[idx]
        elif action == 'DOWN' and row < 2:
            new_state[idx], new_state[idx+3] = new_state[idx+3], new_state[idx]
        elif action == 'LEFT' and col > 0:
            new_state[idx], new_state[idx-1] = new_state[idx-1], new_state[idx]
        elif action == 'RIGHT' and col < 2:
            new_state[idx], new_state[idx+1] = new_state[idx+1], new_state[idx]
        else:
            return None # Hành động không hợp lệ
        
        return tuple(new_state)

    def solve_bfs(self):
        node = Node(self.initial_state)
        if node.state == self.goal_state:
            return self.solution(node)

        # Frontier: hàng đợi FIFO
        frontier = deque([node])
        # Reached: các trạng thái đã thăm
        reached = {self.initial_state}

        while frontier:
            node = frontier.popleft()
            
            for action in self.actions:
                child_state = self.result(node.state, action)
                
                if child_state and child_state not in reached:
                    child_node = Node(child_state, node, action)
                    if child_state == self.goal_state:
                        return self.solution(child_node)
                    
                    reached.add(child_state)
                    frontier.append(child_node)
        
        return None

    def solution(self, node):
        path = []
        while node.parent:
            path.append((node.action, node.state))
            node = node.parent
        return path[::-1]

def print_board(state):
    for i in range(0, 9, 3):
        print(state[i:i+3])
    print("-----------")

start = [0, 2, 3,
         1, 8, 5,
         4, 7, 6]

goal = [1, 2, 3,
        4, 5, 6,
        7, 8, 0]

print("Trạng thái bắt đầu:")
print_board(start)

agent = EightPuzzleAgent(start, goal)
result_path = agent.solve_bfs()

if result_path:
    print(f"Tìm thấy lời giải sau {len(result_path)} bước:\n")
    for step, (action, state) in enumerate(result_path, 1):
        print(f"Step {step}: {action}")
        print_board(state)
else:
    print("Không tìm thấy lời giải.")
```
