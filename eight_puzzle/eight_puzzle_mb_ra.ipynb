Giải câu đố eight_puzzle bằng model-based reflex agent

```python
import numpy as np
import copy

GOAL = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
INITIAL_STATE = [[0, 2, 3], [1, 7, 5], [4, 8, 6]]
move_name = ""

# Các trạng thái trước
history = []

def find_zero(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return (i, j)

def get_moves(state):
    x, y = find_zero(state)
    moves = []
    # (dx, dy, name)
    directions = [(-1, 0, "UP"), (1, 0, "DOWN"), (0, -1, "LEFT"), (0, 1, "RIGHT")]
    
    for dx, dy, name in directions:
        nx, ny = x + dx, y + dy
        if 0 <= nx < 3 and 0 <= ny < 3:
            moves.append((nx, ny, name))
    return moves

def execute_move(state, move):
    global move_name
    new_state = copy.deepcopy(state)
    zx, zy = find_zero(new_state)
    nx, ny, name = move
    # Đổi vị trí zero
    new_state[zx][zy], new_state[nx][ny] = new_state[nx][ny], new_state[zx][zy]
    move_name = name
    return new_state

def model_based_reflex_agent(current_state):
    global history
    
    if current_state == GOAL:
        print("Goal Reached!")
        return None

    possible_moves = get_moves(current_state)
    
    # Lọc nước đi
    for move in possible_moves:
        resulting_state = execute_move(current_state, possible_moves[np.random.randint(len(possible_moves))])
        if resulting_state not in history:
            history.append(current_state)
            return resulting_state
            
    # Nếu tất cả nước đi đều đã có trong quá khứ, chọn ngẫu nhiên
    return execute_move(current_state, possible_moves[np.random.randint(len(possible_moves))])

def print_state(state):
    for i in state:
        print(i)

# --- Simulation ---
curr = INITIAL_STATE
for step in range(1000):
    print(f"Step {step}")
    print_state(curr)
    print("-" * 15)
    if curr == GOAL: break
    curr = model_based_reflex_agent(curr)
    print(move_name)
```
